```python
raw_test_data = pd.read_csv("D:\\Workspace\\WAN_2577_Sea Trial_Data_Sets\\WAN2577_SeaTrial_Test_Data_Pre_Process_Remove_ScaleError_rev2.csv")
raw_test_data
```

```python
# delete the unused column
test_df = raw_test_data.drop(columns=['Time', 'Eng_Run', 'BWACS_CL', 'BWACS_Direction', 'CYL1_LW_Temp_PS', 'CYL1_LW_Temp_ES'])
test_df
```

```python
x_train = train_df.drop('CYL1_LW_Temp_Mean', axis=1)
y_train = train_df['CYL1_LW_Temp_Mean']

x_test = test_df.drop('CYL1_LW_Temp_Mean', axis=1)
y_test = test_df['CYL1_LW_Temp_Mean']
```

```python
##### Regression - sklearn #####

from sklearn.ensemble import ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
# from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor 

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
models={
        "Linear Regression":LinearRegression(fit_intercept=False),

        "Ridge Regression":Ridge(alpha=0.1, 
                                 fit_intercept=False),

        "Lasso Regression":Lasso(alpha=10.0, 
                                 fit_intercept=False),

        "Elastic Net Regression":ElasticNet(alpha=0.1, 
                                            fit_intercept=False, 
                                            l1_ratio=0.5),

        # "Polynomial Regression":PolynomialFeatures(),

        "Extra Trees Regression":ExtraTreesRegressor(max_depth=5,
                                                     min_samples_leaf=3,
                                                     min_samples_split=2,
                                                     n_estimators=100),

        "Random Forest Regression": RandomForestRegressor(max_depth=7,
                                                          min_samples_leaf=5,
                                                          min_samples_split=5,
                                                          n_estimators=100),

        "kNN Regression(n_neighbors=3)": KNeighborsRegressor(n_neighbors=3,
                                                             weights="distance"),

        "XGBOOST Regression":XGBRegressor(gamma=0.1,
                                          learning_rate=0.5,
                                          max_depth=3,
                                          n_estimators=200,
                                          objective='reg:squarederror', 
                                          tree_method='gpu_hist', 
                                          gpu_id=0),  # Added GPU support.

        "CATBOOST Regression":CatBoostRegressor(depth=5,
                                                iterations=300,
                                                learning_rate=0.1,
                                                verbose=False, 
                                                task_type="GPU", 
                                                devices='0:1'),  # Added GPU support.
        "Light GBM Regression": LGBMRegressor(learning_rate=1.0,
                                              n_estimators=300,
                                              num_leaves=31,
                                              device='gpu')  # Added GPU support.
        }
```

```python
# 각 모델의 성능 지표를 저장할 딕셔너리 생성
model_metrics = {}

# 각 모델 학습 및 평가
for name, model in models.items():
    # 모델 학습
    model.fit(x_train, y_train)
    
    # 예측값 생성
    y_preds = model.predict(x_test)
    
    def plot_history(y_test, y_preds):
        x = np.arange(len(y_test))
        
        # 1차 선형 회귀
        coefficients = np.polyfit(x, y_test, 1)
        regression_line = np.poly1d(coefficients)

        plt.figure(figsize=(10, 15))  # figsize 인자 조정

        # 첫 번째 subplot
        plt.subplot(2,1,1)
        plt.plot(x, y_test, 'bo', ms=5, label='True Values')
        plt.plot(x, y_preds, 'ro', ms=2, label='Predicted Values')
        plt.plot(x, regression_line(x), 'r-', label='Regression Line')
        plt.xlabel('Index')
        plt.ylabel('Value')
        plt.title('Regression Line')
        plt.legend()

        # 두 번째 subplot
        plt.subplot(2,1,2)
        # 데이터 최대, 최소값 구하기
        data_max_x = max(y_test)
        data_min_x = min(y_test)
        data_max_y = max(y_preds)
        data_min_y = min(y_preds)

        # 그래프 영역을 데이터 범위에 맞게 설정
        plt.xlim([data_min_x, data_max_x])
        plt.ylim([data_min_y, data_max_y])

        plt.scatter(y_test, y_preds)
        plt.xlabel('True Values [CYL1_LW_Temp_Mean]')
        plt.ylabel('Predictions [CYL1_LW_Temp_Mean]')
        plt.axis('equal')
        plt.axis('square')
        
        # plt.xlim([0, plt.xlim()[1]])
        # plt.ylim([0, plt.ylim()[1]])

        plt.tight_layout()
        plt.show()

        # 평가 지표 계산 (여기서는 MSE 사용)
        r2 = r2_score(y_test, y_preds)
        mse = mean_squared_error(y_test, y_preds)
        mae = mean_absolute_error(y_test, y_preds)
        rmse = np.sqrt(mse)
        rmsle = np.sqrt(np.mean(np.power(np.log1p(y_test)-np.log1p(y_preds), 2)))

        # 결과를 딕셔너리에 저장
        model_metrics[name] = {'R2': r2, 'MSE': mse, 'MAE': mae, 'RMSE': rmse, 'RMSLE': rmsle}
        
        # 교차 검증 수행
        cv_scores = cross_val_score(model, x_train, y_train, cv=5)
        print(f"{name} 모델의 교차 검증 점수: {cv_scores.mean()}")

        # 결과 출력
        print(f"{name} 모델의 R2: {r2}")
        print(f"{name} 모델의 MSE: {mse}")
        print(f"{name} 모델의 MAE: {mae}")
        print(f"{name} 모델의 RMSE: {rmse}")
        print(f"{name} 모델의 RMSLE: {rmsle}")

    # y_test와 y_pred 값을 인자로 전달하여 함수 호출
    plot_history(y_test, y_preds)
    
# 최적의 결과를 찾아 출력
best_model = max(model_metrics, key=lambda x: model_metrics[x]['R2'])

print(f"최적의 모델은 {best_model}이며, 지표는 다음과 같습니다:")
print(f"R2: {model_metrics[best_model]['R2']}")
print(f"MSE: {model_metrics[best_model]['MSE']}")
print(f"MAE: {model_metrics[best_model]['MAE']}")
print(f"RMSE: {model_metrics[best_model]['RMSE']}")
print(f"RMSLE: {model_metrics[best_model]['RMSLE']}")
```
