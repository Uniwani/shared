```python
raw_test_data = pd.read_csv("D:\\Workspace\\WAN_2577_Sea Trial_Data_Sets\\WAN2577_SeaTrial_Test_Data_Pre_Process_Remove_ScaleError_rev2.csv")
raw_test_data
```

```python
# delete the unused column
test_df = raw_test_data.drop(columns=['Time', 'Eng_Run', 'BWACS_CL', 'BWACS_Direction', 'CYL1_LW_Temp_PS', 'CYL1_LW_Temp_ES'])
test_df
```

```python
x_train = train_df.drop('CYL1_LW_Temp_Mean', axis=1)
y_train = train_df['CYL1_LW_Temp_Mean']

x_test = test_df.drop('CYL1_LW_Temp_Mean', axis=1)
y_test = test_df['CYL1_LW_Temp_Mean']
```

```python
import tensorflow as tf
from tensorflow.python.client import device_lib
# Print a list of all available GPUs
device_lib.list_local_devices()

from tensorflow import keras
from keras import layers

import warnings
# Filter out all warnings
warnings.filterwarnings("ignore")
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(171, activation='relu', input_shape=[len(x_train.keys())]),
    layers.Dense(171, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.Adam(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse', 'accuracy'])
  return model

model = build_model()

model.summary()

example_batch = x_train[:10]
example_result = model.predict(example_batch)
example_result

# 에포크가 끝날 때마다 점(.)을 출력해 훈련 진행 과정을 표시
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 200

history = model.fit(
  x_train, y_train,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])

hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(20,15))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [CYL1_LW_Temp_Mean]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$CYL1_LW_Temp_Mean^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.legend()
  plt.show()

plot_history(history)

# Evaluate the model on the test data
results = model.evaluate(x_test, y_test, verbose=2)

# Retrieve the loss, MAE, and MSE from the results
loss = results[0]
mae = results[1]
mse = results[2]

# test dataset을 이용한 'CYL1_Wear_AS' 예측 및 도식화
y_preds = model.predict(x_test).flatten()

plt.scatter(y_test, y_preds)
plt.xlabel('True Values [CYL1_LW_Temp_Mean]')
plt.ylabel('Predictions [CYL1_LW_Temp_Mean]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])

# test dataset을 이용한 'CYL1_Wear_AS' 실제값 vs. 예측값 출력
y_preds = model.predict(x_test)
print("테스트 데이터의 Loss 값: ", loss)
for i in range(5):
    print("%d 번째 테스트 데이터의 실제값: %f" % (i, y_test.iloc[i]))
    print("%d 번째 테스트 데이터의 예측값: %f" % (i, y_preds[i][0]))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

def plot_with_regression_line_and_metrics(x, y_test, y_preds):
    """
    데이터셋의 인덱스를 기준으로 실제 값과 예측 값을 표시하고, 회귀선을 그리며,
    R², MSE, MAE, RMSE 성능 지표를 출력하는 함수
    
    Parameters:
    - x: 데이터셋의 인덱스
    - y_test: 실제 값
    - y_preds: 모델에 의해 예측된 값
    - regression_line: 회귀선을 계산하는 함수, x를 인자로 받아 y 값을 반환    
    """
    plt.figure(figsize=(10, 6))
    
    plt.plot(x, y_test, 'bo', ms=5, label='True Values')
    plt.plot(x, y_preds, 'ro', ms=2, label='Predicted Values')
        
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title('Regression Line with Performance Metrics')
    plt.legend()
    
    # 성능 지표 계산 및 출력
    r2 = r2_score(y_test, y_preds)
    mse = mean_squared_error(y_test, y_preds)
    mae = mean_absolute_error(y_test, y_preds)
    rmse = np.sqrt(mse)  # RMSE는 MSE의 제곱근
    
    print(f"R-squared (R²): {r2:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    
    plt.show()

# 예제 호출 방법은 이전과 동일합니다.
plot_with_regression_line_and_metrics(test_df.index, y_test, y_preds)
```
